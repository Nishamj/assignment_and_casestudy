# 3.Plot a bar graph of male vs female buyers participated in the sales

```python
df['Buyer Gender'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='Buyer Gender',palette=['#d44eae','#654ed4'])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Count',fontsize=15)
plt.xlabel('Buyer Gender',fontsize=14)
plt.grid(linewidth = 1,linestyle = '--')
plt.title('Male vs Female buyers participated in the sales',fontsize=18)

```

# 4.Find the top 5 cars based on their sales price

```python
top_car=df.sort_values(['Sale Price'],ascending=[False])
top_car.head()
```

```python
import numpy as np
print('Top 5 cars based on sales price are\n')
print(pd.DataFrame(top_car.head(),columns=['Make','Model','Sale Price']))
```
